```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["font.family"] = "NanumGothic"
```

```python
df = pd.read_csv("../../data/elderly/pre_processing_data/rate_groupby_region_year.csv")
df.head()
```

```python
print(df.columns)
print(df['노인인구_기관수_비율'].describe())

plt.figure(figsize=(8, 6))
sns.boxplot(
    y=df['노인인구_기관수_비율'],
    width=0.4,
    color='skyblue'
)

plt.title("전체의 '노인인구 대비 요양기관수 비율' 분포", fontsize=16)
plt.ylabel('노인인구 대비 요양기관수 비율', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
print(df[df['년도'] == 2024]['노인인구_기관수_비율'].describe())

plt.figure(figsize=(8, 6))
sns.boxplot(
    y=df[df['년도'] == 2024]['노인인구_기관수_비율'],
    width=0.4,
    color='mediumseagreen'
)

plt.title("전체의 '노인인구 대비 요양기관수 비율' 분포", fontsize=16)
plt.ylabel('노인인구 대비 요양기관수 비율', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
print(df['시도명'].unique())
```

```python
print(f"관측 기간: {df['년도'].min()}년 ~ {df['년도'].max()}년")
```

```python
regional_avg = df.groupby('시도명')['노인인구_기관수_비율'].mean().sort_values(ascending=False)
print(regional_avg)

plt.figure(figsize=(12, 6))

sns.barplot(
    x=regional_avg.index,
    y=regional_avg.values,
    palette='tab20'
)

plt.title('시도별 노인인구 대비 요양기관 수 평균 비율 비교', fontsize=16)
plt.xlabel('시도명', fontsize=12)
plt.ylabel('노인인구 대비 요양기관 수 평균 비율', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
regional_max = df.groupby('시도명')['노인인구_기관수_비율'].max().sort_values(ascending=False)
print(regional_max)

plt.figure(figsize=(12, 6))

sns.barplot(
    x=regional_max.index,
    y=regional_max.values,
    palette='tab20'
)

plt.title('시도별 노인인구 대비 요양기관 수 최대 비율 비교', fontsize=16)
plt.xlabel('시도명', fontsize=12)
plt.ylabel('노인인구 대비 요양기관 수 최대 비율', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
yearly_avg = df.groupby('년도')['노인인구_기관수_비율'].mean()
print(yearly_avg)

plt.figure(figsize=(10, 6))
sns.lineplot(
    x=yearly_avg.index,
    y=yearly_avg.values,
    marker='o',
    color='darkblue',
    linewidth=2.5
)

plt.title('평균 노인인구 대비 요양기관 수 비율의 연도별 변화', fontsize=16)
plt.xlabel('년도', fontsize=12)
plt.ylabel('노인인구 대비 요양기관 수 평균 비율', fontsize=12)

# X축 눈금 설정 (정수 연도만 표시)
plt.xticks(yearly_avg.index, rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)

# 4. 그래프 출력
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))

num_regions = df['시도명'].nunique()
palette = sns.color_palette("tab20", n_colors=num_regions)

sns.lineplot(
    data=df,
    x='년도',
    y='노인인구_기관수_비율',
    hue='시도명',
    palette=palette,
    linewidth=3,
    alpha=0.8,
    style='시도명',
    markers=True,
    dashes=False
)

plt.title('시도별 노인인구 대비 요양기관 수 비율의 연도별 변화', fontsize=16)
plt.xlabel('년도', fontsize=12)
plt.ylabel('노인인구 기관수 비율', fontsize=12)

plt.xticks(df['년도'].unique(), rotation=45)

plt.legend(title='시도명', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```
